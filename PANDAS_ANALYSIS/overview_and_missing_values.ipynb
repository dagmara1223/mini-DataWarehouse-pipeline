```python
import pandas as pd
```

**1. Introduction**

```python
'''
Main analysis
'''

df = pd.read_csv('Amazon_Sale_Report.csv', index_col='index')
print(f"Rows, columns: {df.shape} ")
df.head(n=3)
```

**2. Data Overview**

```python
df.info()
```

```python
df.describe()
```

As we can see, dataset contains 128975 rows and 23 columns. Several columns have missing values - for example column 'Unnamed' is very suspicious. Let's make a copy of our dataset to ensure that original data will stay untouched.

```python
df_copy = df.copy()
```

**3. Missing values analysis**

```python
# missing values all width - full na row

df_copy.isna().all()
```

The first column to further examination is 'Unnamed' column.

```python
df_copy['Unnamed: 22']
```

```python
unnamed_counts = df_copy['Unnamed: 22'].value_counts()
print(f"Values in 'Unnamed 22' column: {unnamed_counts}")
print("")
unnamed_missing = df_copy['Unnamed: 22'].isna().sum()
print(f"Values missing in column 'Unnamed 22': {unnamed_missing}")
```

As we can see - this column is not truly informative.  <br>
✅ - TO DO: Column 'Unnamed 22' is to be deleted.

```python
df_copy.drop('Unnamed: 22', axis=1, inplace=True)
df_copy.head(3)
```

Let's check other missing values.

```python
df_copy.isnull().sum().sort_values()
missing_values_pr = df_copy.isnull().sum() / len(df_copy) * 100
missing_values_pr.sort_values()
```

As we can see - other columns to further examination are: fulfilled-by, promotion-ids, currency, Amount. But first - let's change all column names to lower case. <br>
✅ - TO DO: change column names to lower

```python
df_copy.columns = df_copy.columns.str.lower()
df_copy.columns
```

Since column 'fulfilled-by' is missing 70% of values and column 'promotion-ids' is missing 40% - I've decided to drop such columns.  <br>
✅ - TO DO : drop 'fulfilled_by', drop: 'promotion-ids'

```python
df_copy.drop('fulfilled-by', axis=1, inplace=True)
df_copy.drop('promotion-ids',axis=1,inplace=True)
df_copy.head(3)
```

Further 'currency' column analysis

```python
df_copy['currency'].value_counts()
df_copy['currency']
```

As we can see - 'INR' is the only value in such column. I've decided to fill null values using 'INR'. <br>
- ✅ TO DO : Fill null values in 'currency' column with 'INR'

```python
df_copy['currency'] = df_copy['currency'].fillna('INR')
```

```python
# check
df_copy['currency'].isna().sum()
```

Further 'amount' column analysis

```python
df_copy['amount']
```

Since 'amount' colum contains float values - I will fill null values using mean value.  <br>
- ✅ TO DO: fill null 'amount' values using mean

```python
mean_val = df_copy['amount'].mean().round(2)
df_copy['amount'] = df_copy['amount'].fillna(mean_val)
```

```python
#check
df_copy['amount'].isna().sum()
```

Further 'courier status' column analysis

```python
df_copy['courier status'].value_counts()
```

For the courier_status column, we considered two imputation strategies: filling missing values with the mode ('Shipped') or labeling them as 'Unknown'. I've decided to proceed with the latter to maintain data integrity and avoid making assumptions about the logistics process.
- ✅ TO DO: Fill missing 'courier status' values with 'Unknown'

```python
df_copy['courier status'] = df_copy['courier status'].fillna('Unknown')
```

```python
# check
df_copy['courier status'].isna().sum()
```

Now, let's handle missing values in columns:
'ship-city', 'ship-state', 'ship-country','ship-postal-code'

```python
df_copy['ship-city'].value_counts()
```

```python
df_copy['ship-state'].value_counts()
```

Since both ship-city and ship-state exhibit high cardinality, filling missing values with the most frequent category (mode) would significantly distort the geographical distribution. Therefore, we have opted to impute all NaNs with 'Unknown' to maintain the accuracy of our location-based analysis <br>
✅ TO DO
- fill null values in columns 'ship-city', 'ship-state' with 'Unknown'

```python
df_copy[['ship-city','ship-state']]=  df_copy[['ship-city', 'ship-state']].fillna('Unknown')
```

```python
#check
df_copy[['ship-city', 'ship-state']].isna().sum()
```

```python
df_copy['ship-postal-code']
```

During the analysis, we identified that the postal-code column was stored as a float. To ensure data consistency, we converted it to integer format. Regarding missing values, since postal codes are highly specific, we filled them with a placeholder value (0) to distinguish them from valid codes while maintaining a numeric data type <br>
✅ TO DO: retype 'postal-code' to 'int' and fill missing values with 0

```python
# We can not convert values before dealing with null-values.
df_copy['ship-postal-code'] = df_copy['ship-postal-code'].fillna(0.0)

df_copy['ship-postal-code']= df_copy['ship-postal-code'].astype('int')
```

```python
#check
df_copy['ship-postal-code'].isna().sum()
```

```python
df_copy['ship-country']
```

Since in 'ship-country' we only have IN & Null values I've decided to fill missing values with 'IN' <br>
✅ TO DO: Fill missing values in 'ship-country' with 'IN'

```python
df_copy['ship-country'] = df_copy['ship-country'].fillna('IN')
```

```python
#check
df_copy['ship-country'].isna().sum()
```

After this step, out dataset should be null-less :)
