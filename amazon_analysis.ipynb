**3. Missing values analysis**

The first column to further examination is 'Unnamed' column.

```python
df_copy['Unnamed: 22']
```

```python
unnamed_counts = df_copy['Unnamed: 22'].value_counts()
print(f"Values in 'Unnamed 22' column: {unnamed_counts}")
print("")
unnamed_missing = df_copy['Unnamed: 22'].isna().sum()
print(f"Values missing in column 'Unnamed 22': {unnamed_missing}")
```

As we can see - this column is not truly informative.  <br>
✅ - TO DO: Column 'Unnamed 22' is to be deleted.

```python
df_copy.drop('Unnamed: 22', axis=1, inplace=True)
df_copy.head(3)
```
